```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="Age", hue="Outcome", multiple="stack", bins=15)
plt.title("Age Distribution by Diabetes Outcome")
plt.xlabel("Age")
plt.ylabel("Number of Patients")
plt.legend(title="Outcome", labels=["No Diabetes (0)", "Diabetes (1)"])
plt.tight_layout()
plt.show()
```

```python
df["Outcome"].value_counts(normalize=True) * 100
```

```python
total_count = df["Outcome"].count()
print(total_count)
diabetic = df["Outcome"].sum()
print(f"no of diebetic patients:{diabetic}")
non_diabetic_count = total_count - diabetic
print(f"no of non diebetic patients:{non_diabetic_count}")
#cal percentage
diabetic = (diabetic / total_count) * 100
non_diabetic = (non_diabetic_count / total_count) * 100
print(f"percentage of diebetic:{diabetic}")
print(f"percentage of non_diebetic:{non_diabetic}")
```

```python
categories = ['Diabetic', 'Non-Diabetic']
percentages = [diabetic, non_diabetic]

plt.figure(figsize=(6, 4))
plt.bar(categories, percentages, color=['red', 'blue'])
plt.ylabel('Percentage')
plt.title('diabetis distribution of population')
plt.show()
```

```python
df.loc[:, "AgeGroup"] = pd.cut(
    df["Age"],
    bins=[20, 30, 40, 50, 60, 100],
    labels=["20-30", "31-40", "41-50", "51-60", "60+"]
)


age_group_stats = df.groupby("AgeGroup",observed=False)["Outcome"].agg(
    total_count='count',
    diabetic_count='sum'
)


age_group_stats["diabetic_percent"] = (age_group_stats["diabetic_count"] / age_group_stats["total_count"]) * 100


print(age_group_stats)
```

```python
plt.figure(figsize=(8, 5))
colors=["skyblue","red","green","black","yellow"]
plt.bar(age_group_stats.index.astype(str), age_group_stats["diabetic_percent"], color=colors)
plt.title("Diabetes Percentage by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Percentage of Diabetic Patients")
plt.show()
```

```python
df = df.dropna(subset=["BMI"])
```

```python
df.loc[:,"BMI_Category"] = pd.cut(df["BMI"], bins=[0, 18.5, 24.9, 29.9, 34.9, 100],
                            labels=["Underweight", "Normal", "Overweight", "Obese", "Very Obese"])


bmi_group_stats = df.groupby("BMI_Category",observed=False)["Outcome"].agg(
    total_count='count',
    diabetic_count='sum'
)


bmi_group_stats["diabetic_percent"] = (bmi_group_stats["diabetic_count"] / bmi_group_stats["total_count"]) * 100


print(bmi_group_stats)
```

```python
plt.figure(figsize=(8,5))
colors=["red","blue","green","black","brown"]
plt.bar(bmi_group_stats.index.astype(str),bmi_group_stats["diabetic_percent"],color=colors)
plt.title("diabetic percentage by bmi weight")
plt.xlabel("BMIcategory")
plt.ylabel("diabetic percentage")        
plt.show()            
```

```python
correlation = df["Glucose"].corr(df["Outcome"])
print(f"Correlation between Glucose and Outcome: {correlation:.2f}")
subset = df[["Glucose", "Outcome"]].dropna()

# Compute the correlation matrix
corr_matrix = subset.corr()

# Plot the heatmap
plt.figure(figsize=(5, 4))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm",center=0)
plt.title("Correlation Heatmap: Glucose vs Outcome")
plt.show()
```
